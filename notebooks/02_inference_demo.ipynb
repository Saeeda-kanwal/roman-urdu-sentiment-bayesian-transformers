This notebook lets you test the pre-trained BERT-based sentiment model on Roman Urdu sentences.

```python
# Load model and tokenizer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Replace this with your actual model path if uploaded to Hugging Face
model_name = 'Saeeda-Kanwal/roman-urdu-sentiment-model'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Define sentiment labels
labels = ['Negative', 'Neutral', 'Positive']
```

```python
# Run sentiment prediction
sentence = input("Enter a Roman Urdu sentence: ")
inputs = tokenizer(sentence, return_tensors="pt")
outputs = model(**inputs)
pred = torch.argmax(outputs.logits)
print(f"Predicted Sentiment: {labels[pred]}")
```
